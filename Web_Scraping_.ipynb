```
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```
import requests
```

```
import re
```

```
Link = "https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area"
Link_text = requests.get(Link).text
print(Link_text)
```

```
soup = BeautifulSoup(Link_text, 'lxml')
print(soup)
```

```
print(soup.prettify())
```

```
print(soup.title)
```

```
print(soup.title.string)
```

```
soup.a
```

```
soup.find_all('a')
```

```
all_table = soup.find_all('table')
print(all_table)
```

```
our_table = soup.find('table', class_= "wikitable sortable")
```

```
table_links = our_table.find_all('a')
print(table_links)
```

```
list_country = []
for links in table_links:
  list_country.append(links.get('title'))
print(list_country)
```

```
no_country = []
for i in range (len(list_country)):
   no_country.append(i)
```

```
country_no = []
for No,country in zip((no_country),list_country):
  country_no.append((str(No),country))
```
